```python
import re
```

```python
data = pd.read_csv('../datas/bottins.csv', encoding = "ISO-8859-1")
```

```python
text = []
ents = []
```

```python
for row in data.iloc[:, 0]:
    print(row)
    entites = re.findall(r'<(.*?)>', row)
    # print(entites) ---> ['PER', '/PER', 'ACT', '/ACT', 'LOC', '/LOC', 'CARDINAL', '/CARDINAL', 'TITRE', '/TITRE']
    entites = [entite.strip() for entite in entites]

    texte_original = re.sub(r'<.*?>', '', row)
    texte_original = texte_original.strip()

    text.append(texte_original)
    ents.append(entites)
```

```python
with open('bottins.txt', 'r', encoding='utf-8') as file:
    text = file.read()
```

```python
pattern = r'<PER>(.*?)<\/PER>, <ACT>(.*?)<\/ACT>, <LOC>(.*?)<\/LOC>, <CARDINAL>(\d+)<\/CARDINAL>'
```

```python
import re
```

```python
matches = re.findall(pattern, text)
```

```python
matches
```

```python
data = []
```

```python
for match in matches:
    ent = {
        "text": f"{match[0]} {match[1]} {match[2]}",
        "entities": [
            {"start": 0, "end": len(match[0]), "label": "PER"},
            {"start": len(match[0])+1, "end": len(match[0])+len(match[1])+1, "label": "ACT"},
            {"start": len(match[0])+len(match[1])+2, "end": len(match[0])+len(match[1])+len(match[2])+2, "label": "LOC"}
        ]
    }
    data.append(ent)
```
